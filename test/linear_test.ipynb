```python
import numpy as np
import matplotlib.pyplot as plt
from algorithms.regressions.linear_regression import LinearRegression
from algorithms.utils import plot_various_lines
```

```python
leng = 100
X = np.arange(1, leng + 1)
X_points = X
Y = 100 + 2*X  + np.random.normal(0, 10, leng)
X = np.reshape(X, (X.shape[0], 1))
```

```python
model = LinearRegression()
model.fit(X,Y, epochs = 3000,lr= 0.0001)
```

```python
model.w
```

```python
plot_various_lines([(X_points, Y), (X_points, model.predict(X))], ['ro', 'b-'], ['data', 'model'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Step 1: Generate the data
w0 = np.linspace(-10, 10, 100)
w1 = np.linspace(-10, 10, 100)
W0, W1 = np.meshgrid(w0, w1)
z = np.zeros_like(W0)

# Calculate the cost function
for i in range(leng):
    z += (Y[i] - W0 - W1 * X_points[i]) ** 2
z /= (2 * leng)

# Step 2: Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Step 3: Plot the surface
surf = ax.plot_surface(W0, W1, z, cmap='viridis', alpha=0.8)

# Step 4: Add contour plot
contour = ax.contour(W0, W1, z, levels=np.linspace(z.min(), z.max(), 10), colors='black', linestyles="dashed")

# Customize the plot
ax.set_title('3D Surface Plot of Cost Function')
ax.set_xlabel('W0')
ax.set_ylabel('W1')
ax.set_zlabel('Cost')

# Add a color bar which maps values to colors
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Step 5: Display the plot
plt.show()
```
