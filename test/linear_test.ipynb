```python
import numpy as np
import matplotlib.pyplot as plt
from algorithms.regressions.linear_regression import LinearRegression
from algorithms.utils import plot_various_lines
```

```python
leng = 100
X = np.arange(1, leng + 1)
X_points = X
Y = 100 + 2*X  + np.random.normal(0, 10, leng)
X = np.reshape(X, (X.shape[0], 1))
```

```python
model = LinearRegression()
losses = model.fit(X,Y, epochs = 3000,lr= 0.0001)
```

```python
model.w
```

```python
plot_various_lines([(X_points, Y), (X_points, model.predict(X))], ['ro', 'b-'], ['data', 'model'])
```

```python
plt.plot(range(1, 3001,50), model.losses[::50], 'ro')
```

## Testing with various epoch numbers

```python
%matplotlib inline
for i in range(0, 10, 1):
    test_model = LinearRegression()
    test_model.fit(X,Y, epochs = i,lr= 0.000001)
    plot_various_lines([(X_points, Y), (X_points, test_model.predict(X))], ['ro', 'b-'], ['data', 'model'])
    plt.show()
    plt.pause(0.1)
    plt.clf()
```

## Visualizing loss 3d plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Step 1: Define the range for W0 and W1
W0_values = np.linspace(-10, 10, 100)
W1_values = np.linspace(-10, 10, 100)

# Step 2: Calculate the cost function for each pair of W0 and W1
cost_values = []
for W0 in W0_values:
    for W1 in W1_values:
        cost = np.sum((Y - (W0 + W1 * X_points)) ** 2) / 2
        cost_values.append([W0, W1, cost])

# Convert cost_values to a numpy array for easier manipulation
cost_values = np.array(cost_values)

# Step 3: Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Step 4: Plot the surface
ax.plot_trisurf(cost_values[:, 0], cost_values[:, 1], cost_values[:, 2], cmap='viridis', alpha=0.8)

# Customize the plot
ax.set_title('3D Surface Plot of Cost Function')
ax.set_xlabel('W0')
ax.set_ylabel('W1')
ax.set_zlabel('Cost')

# Add a color bar which maps values to colors
m = plt.cm.ScalarMappable(cmap='viridis')
m.set_array(cost_values[:, 2])
fig.colorbar(m, ax=ax, shrink=0.5, aspect=5)

# Step 5: Display the plot
plt.show()
```
